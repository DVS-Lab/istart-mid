```python
import os 
import pandas as pd
from natsort import natsorted
import re
fsldir=('/data/projects/istart-mid/derivatives/fsl')
```

```python
subs=natsorted([d_name for d_name in os.listdir(fsldir) if d_name.startswith('sub') ])

cope_list=natsorted([re.search('(.*).feat',cope).group(1) for cope
                     in os.listdir(
    '/data/projects/istart-mid/derivatives/fsl/sub-1001/L2_task-mid_model-1_type-act_sm-5.gfeat/')
                    if cope.startswith('cope')])
for cope in cope_list:
    oneR_subs=[]
    twoR_subs=[]
    for sub in subs:
        if [d_name for d_name in os.listdir(fsldir+'/%s'%(sub)) if d_name.startswith('L2')]:
            dataname='L2_task-mid_model-1_type-act_sm-5.gfeat/%s.feat/stats/cope1.nii.gz'%(cope)
            if os.path.exists(fsldir+'/%s/%s'%(sub,dataname)):
                twoR_subs.append(fsldir+'/%s/%s'%(sub,dataname))
        else:
            dataname='L1_task-mid_model-1_type-act_run-1_sm-5.feat/stats/%s.nii.gz'%(cope)
            if os.path.exists(fsldir+'/%s/%s'%(sub,dataname)):
                oneR_subs.append(fsldir+'/%s/%s'%(sub,dataname))
    fnames=natsorted(oneR_subs + twoR_subs)
    df=pd.DataFrame(data=fnames,columns=['fnames'])
    df.to_csv('/data/projects/istart-mid/code/L3_data_%s.tsv'%(cope),sep='\t',index=False)
    
```
