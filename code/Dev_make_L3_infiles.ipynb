```python
import os 
import pandas as pd
from natsort import natsorted
import re
fsldir=('/data/projects/istart-mid/derivatives/fsl')
```

```python
'/data/projects/istart-mid/derivatives/fsl/sub-1001/L2_task-mid_model-2_type-ppi_seed-NAcc.gfeat/'
```

```python
subs=natsorted([d_name for d_name in os.listdir(fsldir) if d_name.startswith('sub') ])
print("subs total: %s"%(len(subs)))

cope_list=natsorted([re.search('(.*).feat',cope).group(1) for cope
                     in os.listdir(
    '/data/projects/istart-mid/derivatives/fsl/sub-1001/L2_task-mid_model-1_type-act.gfeat/')
                    if cope.startswith('cope')])
print("number of copes: %s"%(len(cope_list)))
for cope in cope_list:
    print(cope)
    oneR_subs=[]
    twoR_subs=[]
    for sub in subs:
        
        datanameL2='L2_task-mid_model-1_type-act.gfeat/%s.feat/stats/cope1.nii.gz'%(cope)
        datanameL1='L1_task-mid_model-1_type-act_run-1.feat/stats/%s.nii.gz'%(cope)
        if os.path.exists(fsldir+'/%s/%s'%(sub,datanameL2)):
            twoR_subs.append(fsldir+'/%s/%s'%(sub,datanameL2))
        
        elif os.path.exists(fsldir+'/%s/%s'%(sub,datanameL1)):
            oneR_subs.append(fsldir+'/%s/%s'%(sub,datanameL1))
        else:
            print("%s doesn't have recognizale output"%(sub))
    fnames=natsorted(oneR_subs + twoR_subs)
    df=pd.DataFrame(data=fnames,columns=['fnames'])
    df.to_csv('/data/projects/istart-mid/code/L3_data_files/L3_model-1_type-act_Inputdata_%s.tsv'%(cope),
              sep='\t',
              index=False)
display(df)

```

```python
cope_list=natsorted([re.search('(.*).feat',cope).group(1) for cope
                     in os.listdir(
    '/data/projects/istart-mid/derivatives/fsl/sub-1001/L2_task-mid_model-1_type-ppi_seed-NAcc.gfeat/')
                    if cope.startswith('cope')])

for cope in cope_list:
    oneR_subs=[]
    twoR_subs=[]
    for sub in subs:
        if [d_name for d_name in os.listdir(fsldir+'/%s'%(sub)) if d_name.startswith('L2')]:
            dataname='L2_task-mid_model-1_type-ppi_seed-NAcc.gfeat/%s.feat/stats/cope1.nii.gz'%(cope)
            if os.path.exists(fsldir+'/%s/%s'%(sub,dataname)):
                twoR_subs.append(fsldir+'/%s/%s'%(sub,dataname))
        else:
            dataname='L1_task-mid_model-1_type-ppi_seed-NAcc_run-1_sm-.feat/stats/%s.nii.gz'%(cope)
            if os.path.exists(fsldir+'/%s/%s'%(sub,dataname)):
                oneR_subs.append(fsldir+'/%s/%s'%(sub,dataname))
    fnames=natsorted(oneR_subs + twoR_subs)
    df=pd.DataFrame(data=fnames,columns=['fnames'])
    df.to_csv('/data/projects/istart-mid/code/L3_data_files/L3_model-1_type-ppi_Inputdata_%s.tsv'%(cope),sep='\t',index=False)
    
```
